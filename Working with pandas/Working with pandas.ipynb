```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Reading CSV file using Pandas read_csv("Input file location")
data=pd.read_csv("Iris.csv")
#print available plot styles
print(plt.style.available)
plt.style.use('ggplot')
```

```python
#to print the first five data points
data.head()
```

```python
#give info about the data and also whether there is NaN value fields present in the data
data.info()
```

```python
#describe is used to print statistical data such as max,min,count,mean,std
data.describe()
```

```python
#pd.plotting.scatter_matrix:

#green: sentosa and red: abnormal
#c: color
#figsize: figure size
#diagonal: histohram of each features
#alpha: opacity
#s: size of marker
#marker: marker type

color_list=['red' if i=='Iris-setosa' else 'blue' if i=='Iris-versicolor' else 'green' for i in data.loc[:,'Species']]
pd.plotting.scatter_matrix(data.loc[:,data.columns!='Species'],c=color_list,figsize=[15,15],diagonal='hist',alpha=0.5,s=200,marker='+',edgecolor='black')
plt.show()
```

```python
#using seaborn library
#using countplot() to count the number of classes

sns.countplot(x="Species",data=data)
```
